```python
from qiskit import*
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
unknownNumber = '100110111'
```

```python
circuit = QuantumCircuit(len(unknownNumber)+1,len(unknownNumber))
```

```python
circuit.h(range(len(unknownNumber))) #hedamart içinde secret number dizisi oluşturduk 
circuit.x(len(unknownNumber))
circuit.h(len(unknownNumber))

circuit.barrier()

#for indez, one in enumerate(reversed(secretNumber)): #da olabilr

for index, one in enumerate(unknownNumber[::-1]): # secretnumber içinde şndexlenmiş elemanları gösterttik enumerate yardımıyla,[::-1] ile indexleri tersine çevirdik
    print(f"index {index} is {one}")
    
    if one=='1':
        circuit.cx(index,len(unknownNumber)) # sonuncu qubite bağladık
circuit.barrier()
circuit.h(range(len(unknownNumber)))
circuit.barrier()
circuit.measure(range(len(unknownNumber)),range(len(unknownNumber)))
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit,backend=simulator,shots=1).result()
counts = result.get_counts()
print(counts)
```
